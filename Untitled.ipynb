```python
# importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
# read excel file
product = pd.read_excel("product qc.xlsx")
product.head(10)
```

```python
# datatypes of df columns
product.dtypes
```

```python
# split date column to month, year column.
product['date'] = pd.to_datetime(product['date'])

product['month'] = product['date'].dt.month
product['year'] = product['date'].dt.year

product['month'] = product['month'].astype(str)
product['year'] = product['year'].astype(str)

product.head(10)
```

```python
# change month values
month_dct = {'1': 'Jan', '2': 'Feb', '3': 'Mar', '4': 'Apr', '5': 'May', '6': 'Jun', 
             '7': 'Jul', '8': 'Aug', '9': 'Sep', '10': 'Oct', '11': 'Nov', '12': 'Dec'}
product['month'] = product['month'].map(month_dct)
```

```python
# change year vlues
orig_values = list(product['year'])
format_values = ["'" + year[2:] for year in orig_values]
product["year"] = format_values
```

```python
# concatenate two columns using .str.cat() method
product['siku'] = product['month'].str.cat(product['year'], sep=' ')
product.head(10)
```

```python
# figsize
sns.set(style='whitegrid')
```

```python
# bar chart bottom part
plt.figure(figsize=(10,6))
bar2 = sns.barplot(x='siku', y='passed_products', data=product, color="mediumblue", label='passed')
# bar chart top part
bar1 = sns.barplot(x='siku', y='defected_products', data=product, color="mediumvioletred", bottom=product['passed_products'], label='defected')
plt.title("Passed VS Defected Poducts")
plt.xlabel("Months")
plt.ylabel("No. of Products")
plt.legend()
plt.show()
```

```python
# line plots
plt.figure(figsize=(10,6))
sns.lineplot(x='siku', y='products_returned', data=product, linestyle='-',marker='o', markersize=8, color='crimson', label='returned')
sns.lineplot(x='siku', y='customer_compliants', data=product, linestyle='-', marker='*',markersize=12, color='rebeccapurple', label='compliants')
plt.xlabel("Date")
plt.ylabel("Number")
plt.legend()
plt.show()
```
